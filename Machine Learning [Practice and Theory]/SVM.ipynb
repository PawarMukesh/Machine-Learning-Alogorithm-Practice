## Model building

```python
## step=1 handling categorical target veriable
```

```python
process_data.Loan_status.value_counts()
```

```python
process_data.Loan_status = process_data.Loan_status.map({'Y':1,'N':0})
```

```python
process_data.Loan_status.value_counts()
```

```python
## step2 = define X & y
X = process_data.drop('Loan_status',axis=1)
y = process_data.Loan_status
```

```python
X
```

```python
y
```

```python
### create training and testing data
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=3)
```

```python
### handling imbalance data
process_data.Loan_status.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
```

```python
X_smote,y_smote = smote.fit_resample(X_train,y_train)
```

```python
from collections import Counter

print('Before smote:',Counter(y_train))

print('After smote:',Counter(y_smote))
```

```python
### support vector classifier model

from sklearn.svm import SVC

svclassifier = SVC()
svclassifier.fit(X_smote,y_smote)
```

```python
### predict output for X_test
y_hat = svclassifier.predict(X_test)
```

```python
y_hat
```

```python
### model evaluation

from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score,classification_report
```

```python
accuracy = accuracy_score(y_test,y_hat)
accuracy
```

```python
### classification of report measures a quality of prediction,tp,tn,fp,fn
print(classification_report(y_test,y_hat))
```

```python
y_hat_train = svclassifier.predict(X_train)
```

```python
print(classification_report(y_train,y_hat_train))
```

```python
pd.crosstab(y_test,y_hat)
```

```python
## f1score
f1 = f1_score(y_test,y_hat)
f1
```

```python
### cheacking cross validation score
from sklearn.model_selection import cross_val_score

score = cross_val_score(svclassifier,X,y,cv=3,scoring='f1')
print(score)

print('cross validation score:',score.mean())
print('score standard deviation:',score.std())
```

```python
### grid-search cv

from sklearn.model_selection import GridSearchCV

### defining parameter

parameter = {'C':[0.1,5,10,50,60,70],
            'gamma':[1,0.1,0.01,0.001,0.001],
            'random_state':list(range(1,20))}

model = SVC()

grid = GridSearchCV(model,parameter,refit=True,verbose=2,scoring='f1',cv=5)

### fitting the model for grid search
grid.fit(X,y)
```

```python
### print best parameter after tuning
print(grid.best_params_)

### print how model looks like after tunning
print(grid.best_estimator_)
```

```python
best = SVC(C=0.1, gamma=1, random_state=1)
```

```python
best.fit(X_smote,y_smote)
```

```python
y_best = best.predict(X_test)
y_best
```

```python
print(classification_report(y_test,y_best))
```

```python
pd.crosstab(y_test,y_best)
```

```python
f1 = f1_score(y_test,y_best)
f1
```
